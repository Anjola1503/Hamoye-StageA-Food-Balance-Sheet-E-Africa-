Check for Missing Values

```python
df.isnull().sum()
```

We can see that the dataset contains missing values . The column Y2014,Y2015,Y2016,Y2017,Y2018 contains 1589,1548,1535,1506,1436 missing values. 
Lets treat the missing values

```python
df[df.Y2014.isnull()]
```

Let us replace the missing values of the Y2014 Column with the Mode value of that entire column.

```python
modeValue = df['Y2014'].mode()
modeValue[0]
```

```python
df.dropna(subset = ['Y2015'], inplace=True)
df.dropna(subset = ['Y2016'], inplace=True)
df.dropna(subset = ['Y2017'], inplace=True)
df.dropna(subset = ['Y2018'], inplace=True)
```

```python
df['Y2014'].fillna(value=modeValue[0], inplace = True)
df.isnull().sum()
```
